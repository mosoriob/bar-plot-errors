```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
```

```python
input_file='HMEC1_tiaca1_09_09_22_simple.csv'
```

```python
data = pandas.read_csv(input_file, header=0, index_col=None)
```

```python
bar_key = pandas.unique(data['solvente_concentracion'].values)
```

```python
legend = ['Control Negativo', 'hexano_500', 'hexano_375', 'hexano_250',
       'hexano_125', 'hexano_50', 'diclorometano_500',
       'diclorometano_375', 'diclorometano_250', 'diclorometano_125',
       'diclorometano_50', 'acetato_de_etilo_500', 'acetato_de_etilo_375',
       'acetato_de_etilo_250', 'acetato_de_etilo_125',
       'acetato_de_etilo_50', 'etanol_500', 'etanol_375', 'etanol_250',
       'etanol_125', 'etanol_50', 'c_solvente_0']
```

```python
means = []
errors= []
x_pos = np.arange(len(bar_key))
```

```python
for bar in bar_key:
    bar_array = data.loc[data['solvente_concentracion'] == bar]
    mean = float(np.mean(bar_array))
    means.append(mean)
    error = float(np.std(bar_array))
    errors.append(error)
```

```python
print(errors)
print(means)
print(x_pos)
```

```python
fig, ax = plt.subplots()
ax.bar(x_pos, means, yerr=errors, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Coefficient of Thermal Expansion ($\degree C^{-1}$)')
ax.set_xticks(x_pos)
ax.set_xticklabels(legend)
ax.set_title('Coefficent of Thermal Expansion (CTE) of Three Metals')
ax.yaxis.grid(True)
plt.xticks(rotation=90)
# Save the figure and show
plt.tight_layout()
plt.savefig('bar_plot_with_error_bars.png')
plt.show()
```
